# Linear Regression

## Optimal Solution is given by : x<sup>*</sup> = (A<sup>T</sup>A)<sup>-1</sup>A<sup>T</sup>y




Below function that gives us the following with variables :<br>A ; A<sup>T</sup><br>Product : (A<sup>T</sup>A)<br>Inverse : (A<sup>T</sup>A)<sup>-1</sup><br>Observation : y<br>Optimal Solution : x<sup>*</sup>

```python
def calc_A_x_star(df,name):
    A=df.loc[:, df.columns != name]
    A=pd.DataFrame(A[:].values)
    A_transpose = A.T
    product=A_transpose.dot(A)
    inverse=pd.DataFrame(np.linalg.inv(product.values))
    y=df[name]
    x_star=pd.DataFrame(inverse.dot(A_transpose).dot(y))
    return [A,x_star]
```

#### Linear regression problem in 2 Dimension

```python
def calc_A_x_star(df,name):
    A=df.loc[:, df.columns != name]
    A=pd.DataFrame(A[:].values)
    A_transpose = A.T
    product=A_transpose.dot(A)
    inverse=pd.DataFrame(np.linalg.inv(product.values))
    y=df[name]
    x_star=pd.DataFrame(inverse.dot(A_transpose).dot(y))
    return [A,x_star]
```

```python
def loss(A,x_star,y):
    A=pd.DataFrame(A[:].values)
    ax_b = A.dot(x_star)[0].values-y.values
    return np.sum(ax_b**2)/(2*len(ax_b))
```
